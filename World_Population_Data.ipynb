```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Top 10 countries based on World Population Percentage**

```python
world_population_data.sort_values(by='World Population Percentage', ascending=False).head(10)
```

**Correlation Matrix**

```python
numeric_data = world_population_data.select_dtypes(include='number')
numeric_data.corr()
```

```python
# Heatmap
sns.heatmap(numeric_data.corr(), annot=True)
plt.show()
```

```python
# increasing the size of the above heatmap
plt.figure(figsize=(10,8))
sns.heatmap(numeric_data.corr(), annot=True)
plt.show()
```

**Mean population by continent and sorting it by 2022 population**

```python
world_population_data.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
```

**Oceania continent records**

```python
world_population_data[world_population_data['Continent'] == 'Oceania']
```

**Continent and mean population chart**

```python
## first we need to create a new dataset with only continent names and mean population values

continent_data = world_population_data.groupby('Continent')[world_population_data.columns[5:13][::-1]].mean(numeric_only=True)
#[5:13] will give us only population values columns
#population columns are from 2022 to 1970. [::-1] will reverse it

plt.rcParams['figure.figsize'] = (20,8)
#increase the size of the plot

continent_data.plot()
```

```python
# transposing data will make it easy to understand trend
continent_data = continent_data.transpose()
continent_data
```

```python
continent_data.plot()
```
